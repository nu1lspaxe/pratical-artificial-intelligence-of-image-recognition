```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('data/s.jpg', 0)
height, width = img1.shape

# img1
img2 = np.zeros((height, width), np.uint8)
img2[100:200, 100:300] = 255  # 
img3 = np.hstack((img1, img2))  # 
plt.imshow(img3, cmap='gray'), plt.title('img1/img2')
plt.xticks([]), plt.yticks([])
plt.show()

# 
img4 = cv2.bitwise_and(img1, img2)  # AND
img5 = cv2.bitwise_or(img1, img2)  # OR
img6 = cv2.bitwise_not(img1)  # NOT
img7 = np.hstack((img4, img5, img6))  # 
plt.imshow(img7, cmap='gray'), plt.title('AND/OR/NOT (gray)')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.imshow(img7, cmap='jet'), plt.title('AND/OR/NOT (jet)')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img7, cmap='spring'), plt.title('AND/OR/NOT (spring)')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img7, cmap='summer'), plt.title('AND/OR/NOT (summer)')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img7, cmap='autumn'), plt.title('AND/OR/NOT (autumn)')
plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img7, cmap='winter'), plt.title('AND/OR/NOT (winter)')
plt.xticks([]), plt.yticks([])
plt.show()
```
